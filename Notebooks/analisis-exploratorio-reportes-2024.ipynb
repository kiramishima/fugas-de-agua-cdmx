```python
import polars as pl
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme()
```

```python
df = pl.read_csv('../DATASETS/reportes_agua_2024_01.csv')
```

# Exploracion de los datos

```python
df.head()
```

```python
df.tail()
```

```python
pl.DataFrame({'columna': df.columns, 'tipo data': df.dtypes}).to_pandas()
```

```python
# Parse/Cast de los datos al tipo de dato correcto
df_cast = df.select(
    pl.col("folio_incidente").cast(pl.Utf8).alias('folio'),
    pl.col("fecha_registro_incidente").str.to_date(),
    pl.col("id_reporte").cast(pl.Utf8),
    pl.col("fecha_reporte").str.to_date(),
    pl.col("hora_reporte").cast(pl.Utf8),
    pl.col("clasificacion").cast(pl.Utf8),
    pl.col("reporte").cast(pl.Utf8).alias("tipo_de_falla"),
    pl.col("medio_recepcion").cast(pl.Utf8),
    pl.col("alcaldia_catalogo").cast(pl.Utf8),
    pl.col("colonia_catalogo").cast(pl.Utf8),
    pl.col("latitud").cast(pl.Float64),
    pl.col("longitud").cast(pl.Float64)
)
df_cast.head()
```

```python
df_cast.null_count()
```

```python
total_records = df_cast.shape[0]

df_nulos = df_cast.select(
    (100 - (pl.col("folio").null_count()/total_records * 100).round(2)).alias('folio'),
    (100 - (pl.col("fecha_registro_incidente").null_count()/total_records * 100).round(2)).alias('fecha_registro_incidente'),
    (100 - (pl.col("id_reporte").null_count()/total_records * 100).round(2)).alias('id_reporte'),
    (100 - (pl.col("fecha_reporte").null_count()/total_records * 100).round(2)).alias('fecha_reporte'),
    (100 - (pl.col("hora_reporte").null_count()/total_records * 100).round(2)).alias('hora_reporte'),
    (100 - (pl.col('clasificacion').null_count()/total_records * 100).round(2)).alias('clasificacion'),
    (100 - (pl.col("tipo_de_falla").null_count()/total_records * 100).round(2)).alias('tipo_de_falla'),
    (100 - (pl.col("medio_recepcion").null_count()/total_records * 100).round(2)).alias('medio_recepcion'),
    (100 - (pl.col("alcaldia_catalogo").null_count()/total_records * 100).round(2)).alias('alcaldia_catalogo'),
    (100 - (pl.col("colonia_catalogo").null_count()/total_records * 100).round(2)).alias('colonia_catalogo'),
    (100 - (pl.col("latitud").null_count()/total_records * 100).round(2)).alias('latitud'),
    (100 - (pl.col("longitud").null_count()/total_records * 100).round(2)).alias('longitud')
)
```

```python
data_nulls = pl.DataFrame({'columns': df_nulos.columns, 'values': df_nulos.to_numpy()[0]})
data_nulls
```

```python
fig, ax = plt.subplots(figsize=(14, 7))
barplot = sns.barplot(data=data_nulls, x='columns', y='values', hue='values', ax=ax, legend=False, palette=['steelblue'])
ax.tick_params(axis='x', rotation=90)
ax.set_xlabel('Columnas', fontsize=15, fontweight="bold")
ax.set_ylabel('% datos', fontsize=15, fontweight="bold")
ax.set_title('Valores nulos por columna/variable', fontsize=20, fontweight="bold")

for i in ax.containers:
    ax.bar_label(i, fmt='%.2f%%', fontsize=10, fontweight="bold")

plt.show()
```

```python
# tipos de falla
df_cast['tipo_de_falla'].unique().to_list()
```

```python
df_cast['fecha_reporte'].dt.year().value_counts()
```

```python
df_cast.select(
    pl.col('fecha_reporte').dt.year(),
    pl.col('tipo_de_falla')
).filter(
    pl.col('tipo_de_falla') == 'Fuga'
).group_by('fecha_reporte').agg(total_fugas=(pl.col('tipo_de_falla').count()))
```

```python
df_por_tipo_de_falla = df_cast['tipo_de_falla'].value_counts(sort=True).to_pandas()
df_por_tipo_de_falla
```

```python
fig, ax = plt.subplots(figsize=(12, 8))

barplot = sns.barplot(data=df_por_tipo_de_falla, x='count', y='tipo_de_falla', ax=ax, legend=False, orient='h')
ax.tick_params(axis='x', rotation=90)
ax.set_xlabel('Total', fontsize=15, fontweight="bold")
ax.set_ylabel('Tipo de Falla reportada', fontsize=15, fontweight="bold")
ax.set_title('Total de tipos de fallas reportadas', fontsize=20, fontweight="bold")

for i in ax.containers:
    ax.bar_label(i, fontsize=10, fontweight="bold")

plt.show()
```

```python
# Quitamos los que son NA
df2 = df_cast.filter(pl.col('tipo_de_falla') != 'NA')
df2['tipo_de_falla'].value_counts().sort(by='count', descending=True).to_pandas()
```

```python
# Solo los casos de Fugas
df_fugas = df2.filter(pl.col('tipo_de_falla') == 'Fuga')
df_fugas.head()
```

```python
df_fugas['fecha_reporte'].dt.year().unique()
```

```python
df_fugas['alcaldia_catalogo'].replace("NA", "Ciudad de México")
```

```python
df_fugas['alcaldia_catalogo'].unique().to_list()
```

```python
# Fugas por año
df_por_año = df_fugas.group_by([pl.col('fecha_reporte').dt.year(), 'alcaldia_catalogo']).agg(pl.col('tipo_de_falla').count().alias('total'))
df_por_año
```

```python
df_por_año = df_por_año.with_columns(alcaldia_catalogo=pl.col("alcaldia_catalogo").replace("NA", "Ciudad de México"))
df_por_año
```

```python
df_por_año.columns
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

g = sns.barplot(data=df_por_año,
    x="alcaldia_catalogo", y="total",
    hue="fecha_reporte", palette="dark", alpha=.6,
    ax=ax
)

g.set(
    xlabel="DELEGACIÓN",
    ylabel="TOTAL",
    title="REPORTES DE FUGAS DE AGUA POR ALCALDIA Y AÑO"
)
plt.legend(title='AÑO', loc='upper left')
plt.xticks(rotation=90)
# g.despine(left=True)
# g.set_axis_labels("DELEGACIÓN", "TOTAL")
# g.legend.set_title("")

plt.show()
```

```python
df_por_medio = df_fugas \
    .group_by([pl.col('fecha_reporte').dt.year(), 'medio_recepcion']) \
    .agg(total_fugas=(pl.col('tipo_de_falla').count())) \
    .filter(pl.col('total_fugas') > 50) \
    .sort(by=['fecha_reporte', 'total_fugas'], descending=True).to_pandas()
```

```python
df_por_medio.columns
```

```python
fig, ax = plt.subplots(figsize=(20, 8))

g = sns.barplot(data=df_por_medio,
    x="medio_recepcion", y="total_fugas",
    hue="fecha_reporte", palette="dark", alpha=.6,
    ax=ax
)

g.set(
    xlabel="DELEGACIÓN",
    ylabel="TOTAL",
    title="REPORTES DE FUGAS DE AGUA POR ALCALDIA Y AÑO"
)
plt.legend(title='AÑO', loc='upper right')
plt.xticks(rotation=90)
# g.despine(left=True)
# g.set_axis_labels("DELEGACIÓN", "TOTAL")
# g.legend.set_title("")

plt.show()
```
