```python
import pandas as pd
import warnings
import matplotlib.pyplot as plt
import seaborn as sns

warnings.filterwarnings('ignore')
```

```python
df1 = pd.read_parquet('../DATASETS/fugas_2021.parquet')
df1.head()
```

```python
df2 = pd.read_parquet('../DATASETS/fugas_2022-2024.parquet')
df2['fecha_registro_incidente'] = pd.to_datetime(df2['fecha_registro_incidente'])
df2['fecha_reporte'] = pd.to_datetime(df2['fecha_reporte'])
df2.head()
```

```python
df3 = pd.DataFrame([])
df3['folio'] = df1['folio']
df3['fecha_registro_incidente'] = None
df3['id_reporte'] = pd.NA
df3['fecha_reporte'] = pd.to_datetime(df1['fecha'])
df3['hora_reporte'] = pd.NA
df3['clasificacion'] = 'Agua Potable'
df3['tipo_de_falla'] = df1['tipo_de_falla']
df3['medio_recepcion'] = 'Ciudadano (Call Center)'
df3['alcaldia_catalogo'] = df1['alcaldia'].str.upper()
df3['colonia_catalogo'] = df1['colonia_datos_abiertos'].str.upper()
df3['latitud'] = df1['latitud']
df3['longitud'] = df1['longitud']
df3.head()
```

```python
# Join

df_general = pd.concat([df3, df2])
```
